# Rotation de coordonnées avec Numpy

La fonction `np.dot()` est utilisée pour réaliser le produit scalaire entre deux tableaux (ou matrices). Si nous disposons de deux matrices A et B, le produit scalaire est calculé en multipliant les éléments correspondants et en additionnant ces produits.

En ce qui concerne le lien entre le produit scalaire et la rotation de coordonnées, supposons que nous ayons une matrice de rotation R (matrice comporant les angles de rotation souhaités de chacune des coordonnées) et un vecteur de coordonnées V. La multiplication matricielle de R par V donne un nouveau vecteur (une nouvelle matrice) qui est la version tournée de V.

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir les coordonnées des points
points = np.array([[1, 0], [0, 1], [-1, 0], [0, -1]])

# Angle de rotation en radians
angle = np.radians(45)

# Matrice de rotation 2D
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                            [np.sin(angle), np.cos(angle)]])

# Appliquer la rotation aux points
rotated_points = np.dot(points, rotation_matrix)

print("AVANT ROTATION :\n", points)
print("APRES ROTATION :\n", rotated_points)

# Tracer les points avant et après la rotation
plt.scatter(points[:, 0], points[:, 1], label='Avant rotation')
plt.scatter(rotated_points[:, 0], rotated_points[:, 1], label='Après rotation')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title("Rotation de coordonnées")
plt.axis('equal')
plt.show()
```
